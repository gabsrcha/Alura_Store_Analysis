```python
loja['Data da Compra'] = pd.to_datetime(loja['Data da Compra'], dayfirst=True)
loja
```

```python
loja['SAFRA'] = loja['Data da Compra'].dt.to_period('M')
loja
```

```python
faturamento_diario = loja.groupby('SAFRA')['Preço'].sum().reset_index()
faturamento_diario
```

```python
fatur_vend = pd.crosstab(loja['SAFRA'], loja['Vendedor'], values = loja['Preço'], aggfunc= 'sum', margins = True, margins_name = 'Total')
fatur_vend

```

```python
total = loja['Preço'].sum()
print(total)
```

```python
faturamento_lojas = []

for i, loj in enumerate([loja, loja2, loja3, loja4], 1):
    faturamento = loj['Preço'].sum()
    faturamento_lojas.append(faturamento)
    print(f"Faturamento da loja {i}: R${faturamento:.2f}")

faturamento_total = sum(faturamento_lojas)
print(f"Faturamento total: R${faturamento_total:.2f}")
maior_faturamento = max(faturamento_lojas)
menor_faturamento = min(faturamento_lojas)

loja_mais = faturamento_lojas.index(maior_faturamento) + 1
loja_menos = faturamento_lojas.index(menor_faturamento) + 1

print(f"🏆 Loja que mais faturou: Loja {loja_mais} (R${maior_faturamento:.2f})")
print(f"📉 Loja que menos faturou: Loja {loja_menos} (R${menor_faturamento:.2f})")
```

# 2. Vendas por Categoria


```python
for i, loj in enumerate([loja, loja2, loja3, loja4], 1):
    vendascateg = loj.groupby('Categoria do Produto')['Preço'].sum()

    globals()[f'vendascateg{i}'] = vendascateg

    print(f"\nVendas por categoria da loja {i}:\n{vendascateg}\n")

    categoria_mais = vendascateg.idxmax()
    valor_mais = vendascateg.max()

    categoria_menos = vendascateg.idxmin()
    valor_menos = vendascateg.min()

    print(f"➡️ Categoria que mais faturou na loja {i}: {categoria_mais} (R${valor_mais:.2f})")
    print(f"⬅️ Categoria que menos faturou na loja {i}: {categoria_menos} (R${valor_menos:.2f})")
```

# 3. Média de Avaliação das Lojas

```python
av_loja = []
for i, loj in enumerate([loja, loja2, loja3, loja4], 1):
    media_avaliacao = loj['Avaliação da compra'].mean()
    print(f"Média de avaliação da loja {i}: {media_avaliacao:.2f}")
    av_loja.append(media_avaliacao)

l_av = min(av_loja)
b_av = max(av_loja)

loja_l = av_loja.index(l_av) + 1
loja_b = av_loja.index(b_av) + 1

print(f"A loja com a menor média de avaliação é a loja {loja_l} com uma média de {l_av:.2f}")
print(f"A loja com a maior média de avaliação é a loja {loja_b} com uma média de {b_av:.2f}")
```

# 4. Produtos Mais e Menos Vendidos

```python
for i, loj in enumerate([loja, loja2, loja3, loja4], 1):
    vendaprod = loj['Produto'].value_counts()

    globals()[f'vendaprod{i}'] = vendaprod
    print(f"\nProdutos mais vendidos da loja {i}:\n{vendaprod.nlargest(5)}\n")
    print(f"\nProdutos menos vendidos da loja {i}:\n{vendaprod.nsmallest(5)}\n")
    print("-*-*-" *10)
```

# 5. Frete Médio por Loja


```python

frete_medio = []
for i, loj in enumerate([loja, loja2, loja3, loja4], 1):
    media_frete = loj['Frete'].mean()
    frete_medio.append(media_frete)
    print(f"Frete médio da loja {i}: R${media_frete:.2f}")
```

# Gráficos


```python
faturamento_lojas = [loja['Preço'].sum() for loja in [loja, loja2, loja3, loja4]]
nomes_lojas = [f"Loja {i}" for i in range(1, 5)]

df_faturamento = pd.DataFrame({
    'Loja': nomes_lojas,
    'Faturamento': faturamento_lojas
}).sort_values(by='Faturamento')

sns.set_theme(style="whitegrid")

plt.figure(figsize=(10, 6))

ax = sns.barplot(data=df_faturamento, y='Loja', x='Faturamento', palette='Blues_d')

for i, (valor, nome) in enumerate(zip(df_faturamento['Faturamento'], df_faturamento['Loja'])):
    plt.text(valor + max(faturamento_lojas)*0.01, i, f"R${valor:,.0f}", va='baseline', fontsize=6)

loja_menor = df_faturamento.iloc[0]
plt.barh(loja_menor['Loja'], loja_menor['Faturamento'], color='salmon')

plt.title('Faturamento Total por Loja - Análise de Desempenho', fontsize=12)
plt.xlabel('Faturamento Total (R$)')
plt.ylabel('Loja')
plt.tight_layout()
plt.grid(axis='x', linestyle='--', alpha=0.6)

plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

for i, loj in enumerate([loja, loja2, loja3, loja4], 1):
    sns.set_theme(style="whitegrid")

    vendas_sf_categ = loj.groupby('Categoria do Produto')['Preço'].agg('sum').sort_values()

    plt.figure(figsize=(12, 6))
    ax = sns.lineplot(x=vendas_sf_categ.index, y=vendas_sf_categ.values, marker='o', linewidth=2.5, color='steelblue')

    plt.title(f"Análise de Faturamento por Categoria de Produto - Loja {i}", fontsize=14, weight='bold')

    plt.xlabel('Categoria de Produto', fontsize=12)
    plt.ylabel('Faturamento Total (R$)', fontsize=12)
    plt.xticks(rotation=45, ha='right')
    plt.tight_layout()

    # Adiciona valor em cada ponto
    for x, y in zip(vendas_sf_categ.index, vendas_sf_categ.values):
        plt.text(x, y + max(vendas_sf_categ.values)*0.01, f"R${y:,.0f}", ha='center', fontsize=9)

    max_val = vendas_sf_categ.max()
    min_val = vendas_sf_categ.min()
    max_cat = vendas_sf_categ.idxmax()
    min_cat = vendas_sf_categ.idxmin()

    plt.plot(max_cat, max_val, 'go', markersize=10)
    plt.plot(min_cat, min_val, 'ro', markersize=10)

    plt.annotate(f'Maior: R${max_val:,.0f}', xy=(max_cat, max_val),
                xytext=(0, 15), textcoords='offset points',
                ha='center', color='green', fontsize=10, weight='bold')

    plt.annotate(f'Menor: R${min_val:,.0f}', xy=(min_cat, min_val),
                xytext=(0, -25), textcoords='offset points',
                ha='center', color='red', fontsize=10, weight='bold')

    plt.grid(axis='y', linestyle='--', alpha=0.6)
    plt.show()
```

```python
import folium
from folium.plugins import MarkerCluster

m = folium.Map(location=[-14.2, -51.9], zoom_start=4)  # centro do Brasil

cores = ['blue', 'green', 'red', 'purple']

for i, loj in enumerate([loja, loja2, loja3, loja4]):
    cluster = MarkerCluster().add_to(m)
    for _, row in loj.iterrows():
        folium.CircleMarker(
            location=[row['lat'], row['lon']],
            radius=4,
            color=cores[i],
            fill=True,
            fill_opacity=0.6,
            popup=f"Loja {i+1} - R${row['Preço']:.2f}"
        ).add_to(cluster)

m.save("mapa_lojas.html")
```
