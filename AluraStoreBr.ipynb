```python
faturamento_lojas = []

for i, loj in enumerate([loja, loja2, loja3, loja4], 1):
    faturamento = loj['Preço'].sum()
    faturamento_lojas.append(faturamento)
    print(f"Faturamento da loja {i}: R${faturamento:.2f}")

faturamento_total = sum(faturamento_lojas)
print(f"Faturamento total: R${faturamento_total:.2f}")
maior_faturamento = max(faturamento_lojas)
menor_faturamento = min(faturamento_lojas)

loja_mais = faturamento_lojas.index(maior_faturamento) + 1
loja_menos = faturamento_lojas.index(menor_faturamento) + 1

print(f"🏆 Loja que mais faturou: Loja {loja_mais} (R${maior_faturamento:.2f})")
print(f"📉 Loja que menos faturou: Loja {loja_menos} (R${menor_faturamento:.2f})")
```

# 2. Vendas por Categoria


```python
for i, loj in enumerate([loja, loja2, loja3, loja4], 1):
    vendascateg = loj.groupby('Categoria do Produto')['Preço'].sum()

    globals()[f'vendascateg{i}'] = vendascateg

    print(f"\nVendas por categoria da loja {i}:\n{vendascateg}\n")

    categoria_mais = vendascateg.idxmax()
    valor_mais = vendascateg.max()

    categoria_menos = vendascateg.idxmin()
    valor_menos = vendascateg.min()

    print(f"➡️ Categoria que mais faturou na loja {i}: {categoria_mais} (R${valor_mais:.2f})")
    print(f"⬅️ Categoria que menos faturou na loja {i}: {categoria_menos} (R${valor_menos:.2f})")
```

# 3. Média de Avaliação das Lojas

```python
av_loja = []
for i, loj in enumerate([loja, loja2, loja3, loja4], 1):
    media_avaliacao = loj['Avaliação da compra'].mean()
    print(f"Média de avaliação da loja {i}: {media_avaliacao:.2f}")
    av_loja.append(media_avaliacao)

l_av = min(av_loja)
b_av = max(av_loja)

loja_l = av_loja.index(l_av) + 1
loja_b = av_loja.index(b_av) + 1

print(f"A loja com a menor média de avaliação é a loja {loja_l} com uma média de {l_av:.2f}")
print(f"A loja com a maior média de avaliação é a loja {loja_b} com uma média de {b_av:.2f}")
```

# 4. Produtos Mais e Menos Vendidos

```python
for i, loj in enumerate([loja, loja2, loja3, loja4], 1):
    vendaprod = loj['Produto'].value_counts()

    globals()[f'vendaprod{i}'] = vendaprod
    print(f"\nProdutos mais vendidos da loja {i}:\n{vendaprod.nlargest(5)}\n")
    print(f"\nProdutos menos vendidos da loja {i}:\n{vendaprod.nsmallest(5)}\n")
    print("-*-*-" *10)
```

```python
sns.set(style="whitegrid")
lojas = [loja, loja2, loja3, loja4]
fig, axs = plt.subplots(len(lojas), 2, figsize=(16, 14))
fig.suptitle('Comparação de Produtos Mais e Menos Vendidos por Loja', fontsize=18, fontweight='bold')

for i, loj in enumerate(lojas, 1):
    vendaprod = loj['Produto'].value_counts()
    top5 = vendaprod.nlargest(5)
    bottom5 = vendaprod.nsmallest(5)

    sns.barplot(x=top5.values, y=top5.index, palette='Blues_d', ax=axs[i-1, 0])
    axs[i-1, 0].set_title(f'Loja {i} - Top 5 Mais Vendidos')
    axs[i-1, 0].set_xlabel('Qtd Vendida')
    axs[i-1, 0].set_ylabel('Produto')

    sns.barplot(x=bottom5.values, y=bottom5.index, palette='Oranges_d', ax=axs[i-1, 1])
    axs[i-1, 1].set_title(f'Loja {i} - Top 5 Menos Vendidos')
    axs[i-1, 1].set_xlabel('Qtd Vendida')
    axs[i-1, 1].set_ylabel('Produto')

plt.tight_layout(rect=[0, 0, 1, 0.96])
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

for i, loj in enumerate([loja, loja2, loja3, loja4], 1):
    sns.set_theme(style="whitegrid")

    vendas_sf_categ = loj.groupby('Categoria do Produto')['Preço'].agg('sum').sort_values()

    plt.figure(figsize=(12, 6))
    ax = sns.lineplot(x=vendas_sf_categ.index, y=vendas_sf_categ.values, marker='o', linewidth=2.5, color='steelblue')

    plt.title(f"Análise de Faturamento por Categoria de Produto - Loja {i}", fontsize=14, weight='bold')

    plt.xlabel('Categoria de Produto', fontsize=12)
    plt.ylabel('Faturamento Total (R$)', fontsize=12)
    plt.xticks(rotation=45, ha='right')
    plt.tight_layout()

    # Adiciona valor em cada ponto
    for x, y in zip(vendas_sf_categ.index, vendas_sf_categ.values):
        plt.text(x, y + max(vendas_sf_categ.values)*0.01, f"R${y:,.0f}", ha='center', fontsize=9)

    max_val = vendas_sf_categ.max()
    min_val = vendas_sf_categ.min()
    max_cat = vendas_sf_categ.idxmax()
    min_cat = vendas_sf_categ.idxmin()

    plt.plot(max_cat, max_val, 'go', markersize=10)
    plt.plot(min_cat, min_val, 'ro', markersize=10)

    plt.annotate(f'Maior: R${max_val:,.0f}', xy=(max_cat, max_val),
                xytext=(0, 15), textcoords='offset points',
                ha='center', color='green', fontsize=10, weight='bold')

    plt.annotate(f'Menor: R${min_val:,.0f}', xy=(min_cat, min_val),
                xytext=(0, -25), textcoords='offset points',
                ha='center', color='red', fontsize=10, weight='bold')

    plt.grid(axis='y', linestyle='--', alpha=0.6)
    plt.show()
```

# Relatório

## Introdução:

Com base nas análises realizadas sobre as quatro lojas do Senhor João, este relatório tem como objetivo identificar qual loja apresenta o menor desempenho e deve ser vendida. Foram analisados fatores como faturamento total, categorias de produtos, avaliações dos clientes, produtos mais e menos vendidos, e o custo médio de frete. A seguir, os destaques de cada análise e a recomendação final.

----

## Análise e Pontos Considerados

### 1. Faturamento Total

- **Loja 1**: Lidera com o maior faturamento total, impulsionado por categorias populares.
- **Loja 2**: Apresenta um faturamento moderado, mas inferior à Loja 1.
- **Loja 3**: Tem um faturamento baixo, com dificuldades para competir.
- **Loja 4**: Possui o menor faturamento total, sinalizando baixa performance.

### 2. Categorias de Produtos
Nesse caso, todas as lojas têm um comportamento parecido, onde a categoria mais vendida é a de Eletrônicos e a menos vendida sendo a categoria de Livros. Em termos de faturamento e desempenho em todas as categorias, a loja 1 permanece sendo a que vende mais em todas as categorias, comparando-se com as demais. A loja 4 permanece sendo a de menor desempenho nas categorias.

### 3. Avaliações de Clientes
- **Loja 3**: Melhor média de avaliações (**4.05**), indicando alta satisfação dos clientes.
- **Loja 2**: Média de avaliações **4.04**, próxima à Loja 3.
- **Loja 4**: Média de avaliações **4.00**, mostrando desempenho razoável.
- **Loja 1**: Pior média de avaliações (**3.98**), sugerindo desafios na experiência do cliente.

### 4. Produtos Mais e Menos Vendidos
- **Loja 1**: Produtos de alto valor agregado, como móveis e eletrônicos, lideram as vendas.
- **Loja 4**: Produtos mais vendidos são de baixo valor, impactando negativamente o faturamento.

### 5. Frete Médio
- **Loja 1**: Frete mais alto.
- **Loja 4**: Menor custo médio de frete.

---

## Conclusão e Recomendação

A recomendação é vender a **Loja 4** pelos seguintes motivos:
- **Baixo Faturamento**: Apresenta o menor faturamento total.
- **Fraco Desempenho Geral**: Não se destaca em suas vendas.

Por outro lado, a **Loja 1** deve ser mantida, pois é a mais rentável e bem avaliada. As Lojas 2 e 3, embora enfrentem alguns desafios, possuem mais potencial de crescimento do que a Loja 4.
