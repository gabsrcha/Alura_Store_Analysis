```python
import matplotlib.pyplot as plt
import seaborn as sns

vendas_sf_categ = loja.groupby('Categoria do Produto')['Preço'].agg('sum')

plt.figure(figsize=(10, 6))
sns.lineplot(x=vendas_sf_categ.index, y=vendas_sf_categ.values)
plt.title('Vendas por Categoria')
plt.xlabel('Categoria')
plt.ylabel('Vendas')
plt.xticks(rotation=45, ha = 'right')
plt.tight_layout()
plt.show()
```

```python
loja['Data da Compra'] = pd.to_datetime(loja['Data da Compra'], dayfirst=True)
loja
```

```python
loja['SAFRA'] = loja['Data da Compra'].dt.to_period('M')
loja
```

```python
faturamento_diario = loja.groupby('SAFRA')['Preço'].sum().reset_index()
faturamento_diario
```

```python
fatur_vend = pd.crosstab(loja['SAFRA'], loja['Vendedor'], values = loja['Preço'], aggfunc= 'sum', margins = True, margins_name = 'Total')
fatur_vend

```
